# Funktioner i Python

En funktion i Python tar ett eller flera argument, eller tomt argument, och returnerar ett värde.

```python
def dubbla(x):
    ''' Returnerar talet x multiplicerat med 2.'''
    return 2*x

a = 23
b = dubbla(a)
print(a,b)
```

Funktioner är objekt i sig själva, av typ 'function'.

```python
type(dubbla)
```

Variabelrymder (scope): Lokala variabler och ärvda variabler.
Om en variabel skapas inne i en funktionsdefinition så är den lokal och skild från en variabel mad samma namn utanför och är inte heller definierad utanför funktionens scope.
Variabler som skapats utanför nås, och kan ändras, inne i funktioner, men lokala har företräde. Motsvarande gäller namn på argument till funktionen.

```python
def int_c_ggr(x):
    c = 3
    return c*x

def ext_c_ggr(x):
    return c*x

def arg_c_ggr(c,x):
    return c*x

c, x = 4, 5
print(int_c_ggr(7), ext_c_ggr(7), arg_c_ggr(6,7))

```

```python
def walk(kind_of_walk):
    # global en_lokal_variabel # med kodordet global delas variabeln externt
    en_lokal_variabel = 'walk'
    return kind_of_walk + ' ' + en_lokal_variabel

print(walk('silly'))
try: # om kod i try-blocket ger fel utförs kod i except-blocket istället för avbrott
    print(en_lokal_variabel)
except:
    print('Fel: Lokala variabler finns inte utanför sitt scope')
```

```python
def triangelarea(a,b,c):
    ''' Givet koordinater för tre triangelhörn i planet returneras triangelns area.'''
    xa, ya = a
    xb, yb = b
    xc, yc = c
    return abs((xb-xa)*(yc-ya)-(yb-ya)*(xc-xa))/2

area = triangelarea((2,3),(4,3),(2,6))
print(area)
```

En funktion kan returnera flera värden, som då returneras som en tupel.

```python
def ap(x,y):
    return x+y, x*y

print(type(ap(2,3)))

s, p = ap(2,3)
print(s, p)
print(ap(7,8))
```

Klassmetoder är en särskild typ av funktioner.
De är kopplade till en klass/typ och kan användas på objekt genom `objekt.metod()`  
Exempel: Klassen `'str'` har till exempel en metod `.upper()` som ger en given text skriven i versaler.

```python
s = 'apa'
print(s.upper(), s, 'bepa'.upper())
```

## Importera från moduler.  

Python har en hel del funktioner och klasser som alltid är tillgängliga, men man kan importera moduler som utökar sortimentet.
Detta görs med kodordet `import` följt av modulens namn.  
Exempel på populära moduler är `math`, som definierar matematiska konstanter och funktioner, och `time`, som definierar funktioner som hanterar tid. För att använda objekt från en modul skriver man modulens namn som prefix.

Man skapar en modul genom en fil med pythonkod som definierar modulens objekt. Varje fil med pythonkod i mappen du jobber i fungerar alltså som en modul.

```python
import math
print(math.sin(math.pi/6)) # skriv beräknat sinusvärde av pi/6 radianer. (sin(pi/6) = 1/2)
```

```python
import time
print(time.asctime())
```

Om man vill importera en funktion eller annat objekt `<objekt>` från en modul `<modul>` utan att behöva använda modulens namn som prefix kan man ersätta `import <modul>` med `from <modul> import <objekt>`

```python
from math import pi, sin
print(sin(pi/6))
```
