# Dictionaries, typen `dict`

Listor och tupler är ordnade följder av objekt, som kan indexeras med naturliga tal. Det finns ytterligare ett par standardtyper som beskriver uppsättningar (collections) av object: `dict` och `set`.

En `dict`, uppslagsbok, är en motsvarande struktur, men utan ordning och istället indexerad med "nycklar", tänk en låda med etiketterade objekt.
Nycklarna kan vara av olika typ, tex strängar eller tal. Till varje nyckel hör ett värde. Tillsammans är de en post (`item`) 
Man kan skapa en `dict` som:

```python
inköp_dict = {'Mjölk': '2 liter', 'Bröd': '1 limpa', 'Potatis': '2 kg'}
```

En dict kan alltså skapas genom en kommaseparerad lista av poster (`item`s) `<nyckel>: <värde>` inom {}. En tom lista skapas med ´{}´, eller ´dict()´

Liksom en lista kan en dict modifieras, den är 'mutable'

Man kan komma åt värden i en `dict` genom att indexera med nyckeln.

Man kan ändra eller lägga till värden genom att indexera med nyckeln.

```python
print(inköp_dict)
print(inköp_dict['Mjölk'])
inköp_dict['Mjölk'] = '4 liter'
print(inköp_dict)
inköp_dict['Makaroner'] = '1 paket'
print(inköp_dict)
inköp_dict.add = '1 paket'
print(inköp_dict)
      
```

En nyckel måste vara unik, anger man en nyckel upprepade gånger vid konstruktionen är det den senaste som räknas.

```python
a_dict = {'a': 23, 'b': 42, 'a': 1729}
print(a_dict)
```

`dict` används ofta för att beskriva objekt med flera värden.

```python
mp_address_list = [\
    {'name': 'John Cleese', 'property': 'Fawlty Towers', 'town': 'Torquay'}, \
        {'name': 'Eric Idle', 'property': 'The Shrubbery', 'town': 'Greenwich'}, \
        {'name': 'Michael Palin', 'property': 'Governor\'s Palace', 'town': 'Jerusalem'}, \
        {'name': 'Graham Chapman', 'property': 'Main building', 'town': 'Camelot'}, \
        {'name': 'Terry Jones', 'property': 'Creosote Pond', 'town': 'London'},
                  {'name': 'Terry Gilliam', 'property': 'The Studio', 'town': 'Montone'}    ]
for k in mp_address_list:
    print(f'{k['name']} lives in {k['property']}, {k['town']}')
```

Om man vill utöka en `dict` med fler poster (items) från en annan ´dict´ kan man göra det med `.update()`

```python
a_dict = {'roll': 'Brian'}
a_dict.update({'film': 'Life of Brian', 'skådespelare': 'Graham Chapman'})
```

Gör man en for-loop över en ´dict´ kommer loop-variabeln att gå igenom nycklarna. Vill man ha både nyckel och värde loopar man över d.items().

```python
my_dict = {'a':23, 'b': 42}
for k in my_dict:
    print(f'Nyckeln {k} har värdet {my_dict[k]}')
for k, v in my_dict.items():
    print(f'Nyckeln {k} har värdet {v}')
for k, v in {'a':23, 'b': 42}.items():
    print(f'Nyckeln {k} har värdet {v}')
```

En mängd, ´set´, kan man se som en ´dict´ med nycklar utan värde, och definieras av vilka element den har, utan hänsyn till ordning.
Passar bra att använda om man vill selektera på tillhörighet.

```python
print(set(mp_address_list[0]))
print(set([1,2,3,87,3,2,1,87,34]))
```
